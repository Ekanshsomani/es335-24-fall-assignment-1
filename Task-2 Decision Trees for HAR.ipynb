```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
import tsfel
%config InlineBackend.figure_format = 'retina'
```

**Q1. Use Sklearn Library to train Decision Tress. *[1.5 marks]***

```python
def train_test_func(X_train, y_train, X_test, y_test):
    model = DecisionTreeClassifier(random_state = 21)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average = 'weighted')
    rec = recall_score(y_test, y_pred, average = 'weighted')

    mat = confusion_matrix(y_test, y_pred)

    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")

    plt.figure(figsize = (10, 8))
    sns.heatmap(mat, annot = True, fmt = 'd', cmap = 'Blues', cbar = 'False')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

**Part-A. Train a decision tree model using the raw accelerometer data. Report the accuracy, precision, recall and confusion matrix of the model.**

```python
from HAR.MakeDataset import X_train, y_train, X_test,y_test

def flatten_data(data): return data.reshape(data.shape[0], -1)

X_train_raw = flatten_data(X_train)
X_test_raw = flatten_data(X_test)

train_test_func(X_train_raw, y_train, X_test_raw, y_test)
```

**Part-B. Train a decision tree model using the features obtained by TSFEL. Report the accuracy, precision, recall and confusion matrix of the model.**

```python
def extract_features(X):
    cfg = tsfel.get_features_by_domain()

    tsfel_features_list = []

    for sample in X:
        features = tsfel.time_series_features_extractor(cfg, sample, verbose = 0)
        tsfel_features_list.append(features)
    
    tsfel_features_df = pd.concat(tsfel_features_list, axis = 0, ignore_index = True)

    return tsfel_features_df

X_train_tsfel = extract_features(X_train)
X_test_tsfel = extract_features(X_test)
```

```python
print("Dimensions of training data after extracting features: ", X_train_tsfel.shape)
print("Dimensions of test data after extracting features: ", X_test_tsfel.shape)
```

```python
train_test_func(X_train_tsfel, y_train, X_test_tsfel, y_test)
```

**Part C: Train a decision tree model using the features provided in the dataset. Report the accuracy, precision, recall and confusion matrix of the model.**

```python
# Loading the feature names
dir = os.path.join("HAR", "UCI HAR Dataset")

feature_list = pd.read_csv(os.path.join(dir, "features.txt"), sep = '\s+', header=None)[1].tolist()

# Handling duplicate column names
features_renamed = pd.Series(feature_list)
duplicates = features_renamed[features_renamed.duplicated()].unique()

for dup in duplicates:
    dup_indices = features_renamed[features_renamed == dup].index
    for i, idx in enumerate(dup_indices):
        features_renamed[idx] = f"{dup}.{i+1}"
```

```python
X_train_prov = pd.read_csv(os.path.join(dir, "train/X_train.txt"), sep='\s+', header=None, names = features_renamed)
y_train_prov = pd.read_csv(os.path.join(dir, "train/y_train.txt"), sep='\s+', header=None, names = ['Activity'])
X_test_prov = pd.read_csv(os.path.join(dir, "test/X_test.txt"), sep='\s+', header=None, names = features_renamed)
y_test_prov = pd.read_csv(os.path.join(dir, "test/y_test.txt"), sep='\s+', header=None, names = ['Activity'])
```

```python
print(f"X_train_prov shape: {X_train_prov.shape}")
print(f"y_train_prov shape: {y_train_prov.shape}")
print(f"X_test_prov shape: {X_test_prov.shape}")
print(f"y_test_prov shape: {y_test_prov.shape}")
```

```python
train_test_func(X_train_prov, y_train_prov, X_test_prov, y_test_prov)
```

**Q2. Train Decision Tree with varying depths (2-8) using all above 3 methods. Plot the accuracy of the model on test data vs the depth of the tree. *[1 marks]***

```python
def evaluate_depth(X_train, y_train, X_test, y_test, dataset_name):
    accuracies = []
    depths = range(2, 9)

    for depth in depths:
        model = DecisionTreeClassifier(max_depth = depth, random_state = 21)

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        accuracy = accuracy_score(y_test, y_pred)
        accuracies.append(accuracy)
    plt.plot(depths, accuracies, marker = 'o', label = dataset_name)

plt.figure(figsize = (12, 10))

# Raw Acceleratoin Data
evaluate_depth(X_train_raw, y_train, X_test_raw, y_test, "Raw Acceleration Data")

# TSFEL Extracted Features
evaluate_depth(X_train_tsfel, y_train, X_test_tsfel, y_test, "TSFEL Features")

# Provided Features
evaluate_depth(X_train_prov, y_train_prov, X_test_prov, y_test_prov, "Provided Features")

plt.xlabel("Depth")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Depth")
plt.legend()
plt.grid(True)
plt.show()
```
