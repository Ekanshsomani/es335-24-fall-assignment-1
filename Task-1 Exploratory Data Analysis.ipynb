```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('es335-24-fall-assignment-1/HAR/Combined/Test/WALKING_UPSTAIRS/Subject_2.csv')

print(df.head())
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.plot(df['accx'], df['accy'], df['accz'])

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Waveform Plot')
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df['accx'], label='X-axis')
plt.plot(df['accy'], label='Y-axis')
plt.plot(df['accz'], label='Z-axis')

plt.xlabel('Sample Number')
plt.ylabel('Amplitude')
plt.title('Waveform Plot')
plt.legend()
plt.show()
```
